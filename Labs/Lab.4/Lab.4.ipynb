1. Write a "counter" class that can be incremented up to a specified maximum value, will print an error if an attempt is made to increment beyond that value, and allows reseting the counter. 

```python
class counter:
    def __init__(self, start=0, max=5):
        self.value = start
        self.max = max

    def increment(self):
        self.value += 1
        if self.value > self.max:
            print("Counter has reached its maximum value.")
            self.value = 0

# Example usage
m1 = counter()
for _ in range(6):
    m1.increment()
    print(m1.value)
```

3. Implement a class to represent a rectangle, holding the length, width, and $x$ and $y$ coordinates of a corner of the object. Implement functions that compute the area and perimeter of the rectangle. Make all data members private and privide accessors to retrieve values of data members. 

```python
class rectangle:
    def __init__(self, width=1, height=1, x=0, y=0):
        self.__width = width
        self.__height = height
        self.__x = x
        self.__y = y
    
    def width_setter(self, width):
        if width > 0:
            self.__width = width
        else:
            print("Width must be a positive value.")
    
    def width_getter(self):
        return self.__width

    def height_setter(self, height):
        if height > 0:
            self.__height = height
        else:
            print("Height must be a positive value.")
    
    def height_getter(self):
        return self.__height
    
    def coordinates_setter(self, x, y):
        self.__x = x
        self.__y = y
    
    def coordinates_getter(self):
        return (self.__x, self.__y)

    def area(self):
        return self.__width * self.__height

    def perimeter(self):
        return 2 * (self.__width + self.__height)
    

# Example usage
rect = rectangle()
rect.width_setter(4)
print(rect.width_getter())
rect.height_setter(5)
print(rect.height_getter())
rect.coordinates_setter(2, 3)
print(rect.coordinates_getter())
print("Area:", rect.area())
print("Perimeter:", rect.perimeter())
```

10. Copy the `Canvas` class from lecture to in a python file creating a `paint` module. Copy your classes from above into the module and implement paint functions. Implement a `CompoundShape` class. Create a simple drawing demonstrating that all of your classes are working.

```python
class Canvas:
    def __init__(self, width, height):
        self.width = width
        self.height = height
        # Empty canvas is a matrix with element being the "space" character
        self.data = [[' '] * width for i in range(height)]

    def set_pixel(self, row, col, char='*'):
        if 0 <= row < self.height and 0 <= col < self.width:
            self.data[row][col] = char


    def get_pixel(self, row, col):
        return self.data[row][col]
    
    def clear_canvas(self):
        self.data = [[' '] * self.width for i in range(self.height)]
    
    def v_line(self, x, y, w, **kargs):
        for i in range(x,x+w):
            self.set_pixel(i,y, **kargs)

    def h_line(self, x, y, h, **kargs):
        for i in range(y,y+h):
            self.set_pixel(x,i, **kargs)
            
    def line(self, x1, y1, x2, y2, char='*'):
        dx = abs(x2 - x1)
        dy = -abs(y2 - y1)
        sx = 1 if x1 < x2 else -1
        sy = 1 if y1 < y2 else -1
        err = dx + dy

        while True:
            self.set_pixel(y1, x1, char)

            if x1 == x2 and y1 == y2:
                break

            e2 = 2 * err
            if e2 >= dy:
                err += dy
                x1 += sx
            if e2 <= dx:
                err += dx
                y1 += sy

            
    def display(self):
        print("\n".join(["".join(row) for row in self.data]))

class shape:
    def __init__(self, x=0, y=0):
        self.__x = x
        self.__y = y
    def draw(self, canvas):
        raise NotImplementedError
    
class rectangle(shape):
    def __init__(self, width=1, height=1, x=0, y=0):
        self.__width = width
        self.__height = height
        shape.__init__(self, x, y)
    
    def draw(self, canvas):
        x = self._shape__x 
        y = self._shape__y 
        canvas.h_line(y, x, self.__width)
        canvas.h_line(y + self.__height - 1, x, self.__width)
        canvas.v_line(y, x, self.__height)
        canvas.v_line(y, x + self.__width - 1, self.__height)
    
import math

class circle(shape):
    def __init__(self, radius=1, x=0, y=0):
        shape.__init__(self, x, y)
        self._radius = radius

    def draw(self, canvas):
        cx = self._shape__x
        cy = self._shape__y

        for deg in range(0, 360, 5):
            rad = math.radians(deg)
            px = int(cx + self._radius * math.cos(rad))
            py = int(cy + self._radius * math.sin(rad))
            canvas.set_pixel(py, px, '*')  

class triangle(shape):
    def __init__(self, base=1, height=1, x=0, y=0):
        shape.__init__(self, x, y)
        self._base = base
        self._height = height

        # vertices
        self._x1, self._y1 = x, y
        self._x2, self._y2 = x + base, y
        self._x3, self._y3 = x + base // 2, y + height

    def draw(self, canvas):
        # A → B
        canvas.line(self._y1, self._x1, self._y2, self._x2)
        # B → C
        canvas.line(self._y2, self._x2, self._y3, self._x3)
        # C → A
        canvas.line(self._y3, self._x3, self._y1, self._x1)

class CompoundShape(shape):
    def __init__(self, shapes):
        self.shapes = shapes

    def draw(self, canvas):
        for s in self.shapes:
            s.draw(canvas)

canvas = Canvas(40, 20)
rect = rectangle(10, 5, 5, 5)
circ = circle(3, 20, 10)
tri = triangle(8, 4, 20, 5)
compound = CompoundShape([rect, circ, tri])
compound.draw(canvas)
canvas.display()
```

11. Create a `RasterDrawing` class. Demonstrate that you can create a drawing made of several shapes, paint the drawing, modify the drawing, and paint it again. 

```python
class RasterDrawing:
    def __init__(self):
        self.shapes = []

    def add_shape(self, shape):
        self.shapes.append(shape)

    def remove_shape(self, shape):
        if shape in self.shapes:
            self.shapes.remove(shape)

    def draw(self, canvas):
        canvas.clear_canvas()
        for s in self.shapes:
            s.draw(canvas)

canvas = Canvas(40, 20)

# Create shapes
rect = rectangle(10, 5, 5, 5)
circ = circle(3, 20, 10)
tri  = triangle(8, 4, 20, 5)

# Create drawing
drawing = RasterDrawing()
drawing.add_shape(rect)
drawing.add_shape(circ)
drawing.add_shape(tri)

print("Initial drawing:")
drawing.draw(canvas)
canvas.display()

# Modify the drawing
drawing.remove_shape(circ)
rect2 = rectangle(6, 3, 2, 12)
drawing.add_shape(rect2)

print("\nModified drawing:")
drawing.draw(canvas)
canvas.display()
```
